```python
from crewai import Agent, Task, Crew, Process
from dotenv import load_dotenv

load_dotenv()


def build_logistics_crew(products, problem_focus="inventory turnover and delivery route efficiency"):
    product_list = ", ".join(products)

    logistics_analyst = Agent(
        role="Logistics Analyst",
        goal=(
            "Analyze logistics operations, identify inefficiencies, and provide "
            "clear insights about route efficiency, inventory movement, and product-level bottlenecks."
        ),
        backstory=(
            "You are an experienced logistics analyst who has worked with supply chain, "
            "warehouse, inventory, and last-mile delivery data. You specialize in finding "
            "operational inefficiencies and converting raw logistics information into useful insights."
        ),
        verbose=True
    )

    optimization_strategist = Agent(
        role="Optimization Strategist",
        goal=(
            "Create practical optimization strategies based on logistics insights, "
            "with clear recommendations, expected impact, and implementation steps."
        ),
        backstory=(
            "You are a logistics optimization consultant who helps companies reduce delivery costs, "
            "improve inventory turnover, and increase operational efficiency using AI-assisted planning."
        ),
        verbose=True
    )

    analysis_task = Task(
        description=(
            f"Research and analyze the current logistics situation for the following products: {product_list}.\n\n"
            f"Focus area: {problem_focus}.\n\n"
            "Identify:\n"
            "1. Key logistics challenges\n"
            "2. Route efficiency issues\n"
            "3. Inventory turnover risks\n"
            "4. Demand or stocking concerns\n"
            "5. Product-level operational bottlenecks\n\n"
            "Provide structured insights that can be used by the Optimization Strategist."
        ),
        expected_output=(
            "A structured logistics analysis report with product-wise issues, bottlenecks, "
            "risks, and improvement opportunities."
        ),
        agent=logistics_analyst
    )

    strategy_task = Task(
        description=(
            f"Using the Logistics Analyst's findings, create an optimization strategy for: {product_list}.\n\n"
            "Your strategy should include:\n"
            "1. Product-wise optimization recommendations\n"
            "2. Inventory management improvements\n"
            "3. Route or delivery efficiency improvements\n"
            "4. Suggested KPIs to track\n"
            "5. Short-term and long-term action plan\n"
            "6. Expected business impact"
        ),
        expected_output=(
            "A practical logistics optimization strategy with clear actions, KPIs, "
            "and expected business benefits."
        ),
        agent=optimization_strategist,
        context=[analysis_task]
    )

    crew = Crew(
        agents=[logistics_analyst, optimization_strategist],
        tasks=[analysis_task, strategy_task],
        process=Process.sequential,
        verbose=True
    )

    return crew


if __name__ == "__main__":
    products = ["rice", "wheat flour", "cooking oil", "pulses"]

    crew = build_logistics_crew(
        products=products,
        problem_focus="inventory turnover trends and delivery route optimization"
    )

    result = crew.kickoff()
    print(result)
```
